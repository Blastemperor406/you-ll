```python
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import torch
from torch import optim
from sklearn.model_selection import train_test_split
```

```python
ds=pd.read_csv('/Users/yashagarwal/Darsh/hackathon/OPEN Hack/repo_sync_YouCal/inputs.csv')
ds.drop(['Unnamed: 0'],axis=1,inplace=True)
y=ds["Expected Calories"].to_numpy()
X = ds.drop(['Expected Calories'], axis=1).to_numpy()
```

```python
class LinearFunctionModel(nn.Module):
    def __init__(self,in_features=5,h1=12,h2=6,h3=3,out_features=1):
        super().__init__()
        self.fc1=nn.Linear(in_features,h1)
        #self.fc2=nn.Linear(h1,h2)
        #self.fc3=nn.Linear(h2,h3)
        #self.fc3=nn.Linear(h2,out_features)
        self.fc2=nn.Linear(h1,out_features)
    def forward(self,x):
        x=self.fc1(x)
        x=self.fc2(x)
        return x
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)
```

```python
model = LinearFunctionModel()
loss_function = nn.MSELoss()
print(model)
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
num_epochs = 5000
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = loss_function(y_pred, y_train)
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}, loss: {loss.item()}')
```

```python
with torch.no_grad():
    model.eval()
    y_pred = model(X_test)
    loss = loss_function(y_pred, y_test)
    print(f'Final loss: {loss.item()}')
    print(f'Predictions: {y_pred.squeeze()}')
    print(f'Actual: {y_test.squeeze()}')
    print(f'Coefficients: {model.state_dict()}')
```
